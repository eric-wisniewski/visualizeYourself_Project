```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import pyplot
%matplotlib inline

final_data_e = pd.read_csv("vizSelf_eric.csv", index_col=0, parse_dates=True)
final_data_p = pd.read_csv("vizSelf_parent.csv", index_col=0, parse_dates=True)
final_data_e.head()
final_data_p.head()
```

```python
axe = final_data_e.plot.area(figsize=(12,4), subplots=True)
axp = final_data_p.plot.area(figsize=(12,4), subplots=True)
```

```python
axeb = final_data_e.plot.bar(figsize=(12,4), subplots=True)
axpb = final_data_p.plot.bar(figsize=(12,4), subplots=True)
```

```python
axeh = final_data_e.plot.hist(figsize=(12,4), subplots=True)
axph = final_data_p.plot.hist(figsize=(12,4), subplots=True)
```

```python
axed = final_data_e.plot.density(figsize=(12,4), subplots=True)
axpd = final_data_p.plot.density(figsize=(12,4), subplots=True)
```

```python
axebp = final_data_e.plot.box(figsize=(12,4), subplots=True)
axpbp = final_data_p.plot.box(figsize=(12,4), subplots=True)
```
